# Aula prática

Agora que você aprendeu como empacotar um modelo e disponibiliza-lo para uso, use modelos treinados nas aulas anteriores e o sirva como uma aplicação online. Você pode usar FastAPI ou Flask para criação do servidor e definição do endpoint. 

Adicionais mas não obrigatórios:
- Empacote sua aplicação em um container Docker
- Recupere todos os aterfatos do Mlflow ou outro gerenciador.
  - Os aterfatos devem estar versionados!
- Faça testes unitários da sua aplicação
- Faça uma versão para processamento Batch
- Desenho de arquitetura
